```python
x = np.linspace(-5, 5, 100)
y1 = x**2
y2 = 2*x + 1

plt.plot(x, y1, '-g', label='y = x^2')
plt.plot(x, y2, '--r', label='y = 2x + 1')
plt.title('Multiple Line Plot')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0.1, 5, 100)
y1 = np.exp(x)
y2 = np.log(x)

fig, ax1 = plt.subplots()

ax1.plot(x, y1, '-g', label='y = e^x')
ax1.set_xlabel('X-axis')
ax1.set_ylabel('y = e^x', color='g')

ax2 = ax1.twinx()  # Create a second y-axis
ax2.plot(x, y2, '--b', label='y = ln(x)')
ax2.set_ylabel('y = ln(x)', color='b')

plt.title('Multiple Line Charts with Different Axes')
plt.grid(True)
plt.show()
```

## PairGrid

```python
g=sns.PairGrid(iris,hue='species')
g.map(sns.scatterplot)
```

```python
g=sns.PairGrid(data=iris,hue='species')
g.map_diag(sns.boxplot)
g.map_offdiag(sns.scatterplot)
```

```python
g=sns.PairGrid(data=iris,hue='species')
g.map_upper(sns.violinplot)
g.map_lower(sns.scatterplot)
```
