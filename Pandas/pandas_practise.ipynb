```python
# Select specific columns  (Selects only the Name, Age, and Survived columns from the dataframe)
df_subset = df[['name', 'age', 'survived']]
```

```python
df_subset.head()
```

```python
# Filter passengers who are older than 30

df_filtered = df[df['age'] > 39]
df_filtered
# df_filtered.shape
```

```python
# Sort by Age in descending order
df_sorted = df.sort_values(by='age', ascending=False)
df_sorted.head()
```

```python
# Filter the DataFrame to include only passengers who survived
survived = df[df['survived'] == 1]

# Group by 'sex' and count the number of survivors
total_survived_by_gender = survived.groupby('sex').size()

# Print the result
print(total_survived_by_gender)
```

```python
survived.shape
```

```python
# Group by 'Sex' and calculate the mean age for each group

age_by_sex = df.groupby('sex')['age'].mean()
print(age_by_sex)
```

```python
# Group by 'Sex' and calculate the max age for each group

max_age_by_sex = df.groupby('sex')['age'].max()
print(max_age_by_sex)
```

```python
# Group by 'Sex' and calculate the min age for each group

min_age_by_sex = df.groupby('sex')['age'].min()
print(min_age_by_sex)
```

```python
# Group by 'Pclass' and 'Sex' and calculate the mean fare for each group

fare_by_class_sex = df.groupby(['pclass', 'sex'])['fare'].mean()
print(fare_by_class_sex)
```

```python
# Create a new column "FamilySize" as the sum of 'SibSp' and 'Parch'

df['familysize'] = df['sibsp'] + df['parch']
print(df[['sibsp', 'parch', 'familysize']].head())
df.columns
```

```python
# Check for missing values in the entire DataFrame
print(df.isnull().sum())
```

```python
# Calculate the percentage of missing values for each column
missing_percentage = df.isnull().mean() * 100

# Display the percentage of missing values
print(missing_percentage)
```

```python
# Filter and keep only the columns where less than 70% of the values are missing
df_cleaned = df.loc[:, df.isnull().mean() < 0.7]

# Show the cleaned DataFrame
print(df_cleaned.shape)
print(df_cleaned.columns)
```

```python
# Drop columns with any missing values( Drops 7 out of 14 columns)
df_cleaned_new = df.dropna(axis=1)
df_cleaned_new
```

```python
# # Fill missing values with a constant (e.g., 0)
df_filled = df['body'].fillna(0)

# # Fill missing values with the mean of the column
df['age'] = df['age'].fillna(df['age'].mean())

# # Fill missing values with the median of the column
df['fare'] = df['fare'].fillna(df['fare'].median())
```
