# 1.8. Filtering Data:
    * Filter the data to show listings where the price is greater than $100.


```python
df['price_float'] = df['price'].str.replace('\$|,', '', regex=True)
df['price_float'] = df['price_float'].astype(float)

filterd_df = df[df['price_float'] > 100]
filterd_df.head()
```

# 1.9. Sorting Data:

```python
df_shorted = df.sort_values(by='price_float', ascending=False)
df_shorted[['id', 'price', 'price_float']].head()
```

# 2.5. New Feature Creation:
    * Create a new column called 'price_per_accommodates' by dividing 'price' by 'accommodates'.


```python
df['price_per_accommodates'] = df['price_float'] / df['accommodates']

df[['price_float', 'accommodates', 'price_per_accommodates']].head()
```

# 2.7. String Operations:
    * Extract the first word from the 'name' column and create a new column called 'listing_name_first_word'.

```python
df['listing_name_first_word'] = df['name'].str.split(' ', expand=True)[0]
df[['name', 'listing_name_first_word']].head()
```

# 2.8. Descriptive Statistics by Group:
    * Calculate the standard deviation of 'price' for each 'room_type'.

```python
df.groupby('room_type')['price_float'].std()
```

# 2.9. Find Outliers:
    * Identify listings with a price more than 3 standard deviations above the mean price.

```python
filtered_df = df[df['price_float'] > df['price_float'].mean()+(3*df['price_float'].std())]
filtered_df.head()
```

# 2.10. Replace Values:
    * Replace all values in the 'neighbourhood' column where 'neighbourhood' is 'Downtown' with 'Downtown Seattle'.

```python
df['neighbourhood'] = df['neighbourhood'].str.replace('Downtown', 'Downtown Seattle')
df['neighbourhood'].unique()
```

# 2.11. Pivot Table:

```python
df.pivot_table(values='price_float', index='room_type', columns='neighbourhood', aggfunc='mean')
# df.pivot_table(values='price_float', index=['room_type','neighbourhood'], aggfunc='mean')
```
