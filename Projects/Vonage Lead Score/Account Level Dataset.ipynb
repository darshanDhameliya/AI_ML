```python


# Aggregation logic
def combine_records(group):
    return pd.Series({
        "leads": group["leads"].sum(),
        "display_impressions": group["display_impressions"].sum(),
        "clicks": group["clicks"].sum(),
        "click_through_rate": (group["click_through_rate"] * group["display_impressions"]).sum() / group["display_impressions"].sum(),
        "exposure_time_minutes": group["exposure_time_minutes"].sum(),
        "site_visits": group["site_visits"].sum(),
        "exposure_time_per_site_visit_minutes": (group["exposure_time_per_site_visit_minutes"] * group["site_visits"]).sum() / group["site_visits"].sum(),
    })

# Combine records based on account_id
combined_df = madison_logic.groupby("unified_account_id").apply(combine_records).reset_index()

# Display the combined dataframe
combined_df
```
