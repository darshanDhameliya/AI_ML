# Assignement 1 : (Line Chart)
    * Task: 
        Read a CSV file (temperatures.csv) containing data for year and average temperature of months and plot a line chart for any one month.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
temperatures_df = pd.read_csv("datafiles/temperatures.csv")
temperatures_df.sample(5)
```

    * Task: 
        • Display Title of the graph
        • Display X label
        • Display Y label
        • Change The color of the line chart

```python
plt.plot(temperatures_df['YEAR'], temperatures_df['JAN'], color = 'red')
plt.xlabel('Year')
plt.ylabel('Temperature')
```

    * Task: 
        Repeat Task 2 and display line chart of any 2 or 3 months.
        • Display legend() and each line label
        • Apply Different Style on each line like (color,marker,linewidth,linestyle,markersize)
        • Increase the size of the graph

```python
plt.figure(figsize=(10,7))
plt.plot(temperatures_df['YEAR'], temperatures_df['JAN'], label='JAN', color='red', marker='D', markersize=3)
plt.plot(temperatures_df['YEAR'], temperatures_df['FEB'], label='FEB', linewidth=3)
plt.plot(temperatures_df['YEAR'], temperatures_df['MAR'], label='MAR', linestyle='dashdot')
plt.xlabel('Year')
plt.ylabel('Temperature')
plt.legend()
plt.show()
```

# Assignment 2 : (Barplot)
    * Task: 
        Create a bar plot for a single batsman showing their performance over the years (e.g., Rohit Sharma).(season_record.csv)

```python
session_record_df = pd.read_csv('./datafiles/season_record.csv')
session_record_df.sample(5)
```

    * Task: 
        Plot a grouped bar chart comparing two or more batsmen’s performance over the years. Display Batsman name on xlabel and runs on ylabel. (season_record.csv)

```python
# plt.bar(session_record_df['batsman'], session_record_df['2015'])
filter_session_record_df = session_record_df[session_record_df['batsman'].isin(['V Kohli', 'RG Sharma'])]
filter_session_record_df


plt.bar(np.arange(filter_session_record_df.shape[0])-0.2, filter_session_record_df['2015'], width=0.2)
plt.bar(np.arange(filter_session_record_df.shape[0]), filter_session_record_df['2016'], width=0.2)
plt.bar(np.arange(filter_session_record_df.shape[0])+0.2, filter_session_record_df['2017'], width=0.2)

plt.xticks(np.arange(filter_session_record_df.shape[0]), filter_session_record_df['batsman'])

plt.show()
```

# Assignment 3 : (Scatterplot)
    * Task: 
        Create a scatter plot to show the relationship between total bill and tip amount.(Note : Use ‘tips’ dataset from seaborn datasets)
        • Change the color of the data point
        • Change the marker
        • Change the size of the data points based on ‘size’ column
        • Change the opacity of the graph

```python
tips_df = sns.load_dataset('tips')
tips_df.sample(5)
```

```python
plt.scatter(tips_df['total_bill'], tips_df['tip'], s=tips_df['size']*20, color='blue', marker='+', alpha=0.5)
plt.xlabel('Total Bill')
plt.ylabel('Tip')
plt.show()
```

    * Task: 
        Create a scatter plot to show the relationship between total bill and tip amount, and use color and size to represent additional variables (sex and size).

```python
color_map = { 'Male': 'red', 'Female': 'blue'}

plt.scatter(tips_df['total_bill'], tips_df['tip'], s=tips_df['size']*20, color=[ color_map[i] for i in tips_df['sex'] ], marker='+', alpha=0.5)

plt.xlabel('Total Bill')
plt.ylabel('Tip')
plt.show()
```

    * Assignment 4 : PieChart (batter.csv)

        Create a pie chart representing the proportion of total runs contributed by the top 5 batsmen in the dataset. (batter.csv)

        Create 3 Pie charts side by side of top 5 batsman. In First graph display runs,in second graph display strike rate, in third graph display average.

# Assignment 7 : Heatmap
    * Task:
        Create a heatmap showing the total runs scored in each over for the first innings. (overs.csv)

```python
overs_df = pd.read_csv('./datafiles/overs.csv')
overs_df.sample(5)
```

```python
sample_overs_df = overs_df[(overs_df['innings'] == 1) & (overs_df['ballnumber'] <= 6)]

grid = sample_overs_df.pivot_table(index='overs', columns='ballnumber', values='total_run', aggfunc='sum')

plt.imshow(grid)
plt.colorbar()
plt.show()
```

    * Task:
        Create a heatmap showing in 20 overs on each ball how many six runs are hitted.

```python
sample_overs_df = overs_df[(overs_df['non_boundary'] == 1) & (overs_df['batsman_run'] == 6)]


grid = sample_overs_df.pivot_table(index='overs', columns='ballnumber', values='total_run', aggfunc='count')
grid.fillna(0, inplace=True)
plt.imshow(grid)
plt.colorbar()
plt.show()
```

    * Task:
        Create a heatmap showing how many runs a particular batsman (e.g., YBK Jaiswal) scored per over in the first innings.

```python
sample_overs_df = overs_df[(overs_df['batter'] == 'YBK Jaiswal') & (overs_df['innings'] == 1)]

grid = sample_overs_df.pivot_table(index='overs', columns='ballnumber', values='total_run', aggfunc='sum')
grid.fillna(0, inplace=True)
plt.imshow(grid)
plt.colorbar()
plt.show()
```

    * Task:
        Create a heatmap showing the number of extras (wides, leg byes, etc.) conceded per over by the bowling team in the second innings.


```python
sample_overs_df = overs_df[~overs_df['extra_type'].isnull()]

grid = sample_overs_df.pivot_table(index='overs', columns='extra_type', values='total_run', aggfunc='count')
grid.fillna(0, inplace=True)
plt.imshow(grid)
plt.colorbar()
plt.show()
```

    * Task:
        Create a heatmap showing how many wickets fell in each over of both innings.

```python
sample_overs_df = overs_df[overs_df['innings'].isin([1, 2])]

grid = sample_overs_df.pivot_table(index='overs', columns='innings', values='isWicketDelivery', aggfunc='sum')
plt.imshow(grid)
plt.colorbar()
plt.show()
```
